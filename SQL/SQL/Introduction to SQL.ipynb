```python
sql_query = "SELECT emp.*, contact.Address FROM emp LEFT JOIN contact ON emp.EmpID = contact.EmpID"
result = ps.sqldf(sql_query, locals())
```

```python
result
```

```python
sql_query = "SELECT emp.*, contact.Address, contact.Gender FROM emp LEFT JOIN contact ON emp.EmpID = contact.EmpID"
result = ps.sqldf(sql_query, locals())
```

```python
result
```

```python
sql_query = "SELECT emp.*, contact.Gender FROM emp LEFT JOIN contact ON emp.EmpID = contact.EmpID";
result = ps.sqldf(sql_query, locals())
```

```python
result
```

```python
sql_query = "SELECT emp.*, contact.Gender FROM emp RIGHT JOIN contact ON emp.EmpID = contact.EmpID";
result = ps.sqldf(sql_query, locals())
```
