```python
import pandas as pd
import sqlite3 as sql
```

```python
df = pd.read_excel('employee.xlsx')
conn = sql.connect('Employee.db')

c = conn.cursor()
c.execute("CREATE TABLE IF NOT EXISTS emp ({})".format(' ,'.join(df.columns)))
```

```python
for row in df.iterrows():
    sql = "INSERT INTO emp ({}) VALUES ({})".format(' ,'.join(df.columns), ','.join(['?']*len(df.columns)))
    c.execute(sql, tuple(row[1]))
conn.commit()
```
